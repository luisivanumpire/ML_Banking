# Diccionarios

Es una coleccin de elementos, donde cada uno tiene una llave "key" y un valor "value". 

Los diccionarios se pueden crear con llave {} separando con una coma cada par key:value.

```python
dic1 = {} # un diccionario vaco
dic1
```

```python
dicVehiculo = {
  "marca": "Ford",
  "modelo": "Mustang",
  "ao": 1964
}
dicVehiculo
```

```python
dicVehiculo['ao'] # obteniendo el valor de un "Key"
```

```python
dicVehiculo['marca']
```

```python
# Si el key al que accedemos no existe, se aade automticamente.
dicVehiculo['nro asientos'] = "5"
dicVehiculo
```

## Recorriendo los diccionarios

Podemos recorrer de manera similar que a las Tuplas y Listas.

```python
l = list()
for k in dicVehiculo: #imprimir los "Key"
    print(k)
    l.append(k)

l
```

```python
for v in dicVehiculo:
    print(dicVehiculo[v]) #imprimir los valores del Diccionario
```

```python
# Imprime los key y value del diccionario
for k, v in dicVehiculo.items():
    print("Key:" + str(k) + " y el Valor: " + str(v))
```

### clear()

### get()

### items()

Devuelve una lista con los keys y values del diccionario.

```python
dic2
```

```python
itemss = dic2.items()
print(itemss)             #tupla de la lista del diccionario: ([('a', 1), ('b', 2), ('c', 3)])
print(list(itemss))       #[('a', 1), ('b', 2)]
print(list(itemss)[1])  # primer elemento de la 1ra tupla: a
print(list(itemss)[1][0]) # primer elemento de la 1ra tupla: a
```

### update(<obj>)

Los value son actualizados y si alguna key del nuevo diccionario no esta, es aadida.

```python
dic3 = {'a': 1, 'b': 2, 'c': 3}
dic4 = {'a': 0, 'd': 250}
dic4.update(dic3)     # actualizamos "dic3" con "dic4"
print(dic4)
```
